#Referenced Contribution

```python
data=[train,test]

for item in data:
    family_size = []
    is_alone = []
    
    for i in range(len(item)):
        family_size.append(item['Parch'][i]+item['SibSp'][i])
        if family_size[-1] == 0:
            is_alone.append(1)
        else:
            is_alone.append(0)
        
    item['Family_size'] = family_size
    item['Is_alone'] = is_alone

train
```
